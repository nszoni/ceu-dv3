# **1. Bar charts**

Bar charts are one of the  basic chart types provided by Plotly Express. 

We want to visualize the population distribution of USA with the gapminder sample dataset on a bar chart.

How would we get started?

## **Getting** the data

```python
# creating our filtered data frame which contains records for USA only
df_usa = px.data.gapminder().query('country == "United States"')

# show the dataframe
df_usa
```

## Creating our first chart

```python
# creating the bar chart
fig = px.bar(df_usa, x = 'year', y = 'pop')

# rendering the chart
fig.show()
```

## Adding a title

Now that we have made our first chart with Plotly express we can see that Plotly auto-labels our chart with the column names of our data frame but there is no title by default.

How would we need to modify our initial Plotly Express call to add a title?

```python
# adding the title parameter to our initial chart
fig2 = px.bar(df_usa, x = 'year', y = 'pop', 
              title = 'The population of USA over the years')

fig2.show()
```

```python
# you can also update the existing chart and set the title
fig2.update_layout(title_text="Updated title2")
```

## Changing the coloring and labeling

For our second chart let's visualize what was the life expectancy of Asian countries with over 20 million of population in 2007

How would we need to modify the data set we used and what parameter of our px.bar() function would need to be changed?

```python
# changing the filtering on the data set
df_asia = px.data.gapminder().query('continent == "Asia" and pop > 20000000 and year == 2007')

# taking a look at the data set
df_asia.shape
```

```python
# the dataset
df_asia
```

```python
# adding different columns as parameters
fig3 = px.bar(df_asia, x = 'country', y = 'lifeExp')

fig3.show()
```

If we would like to identify the countries with higher population we can add coloring to the chart by the 'pop' column of our dataset.

```python
# adding coloring using a third variable, population
fig3 = px.bar(df_asia, x = 'country', y = 'lifeExp', color = 'pop')

fig3.show()
```

We can see that some column names are in short forms, let's change the labeling so that our chart is more readable.

```python
# changing the labeling with the labels parameter
fig3 = px.bar(df_asia, x = 'country', y = 'lifeExp', 
              color = 'pop', 
              labels = {'lifeExp' : 'Life Expectancy',
                        'country' : 'Country',
                        'pop' : 'Population'}, 
              title = 'Life Expectancy of bigger Asian countries')

fig3.show()
```

We can also pre-declare variables to use inside our Plotly Express calls.

```python
# creating a seperate dictionary for the labels
labels_ = {'lifeExp' : 'Life Expectancy',
          'country' : 'Country',
          'pop' : 'Population'}
          
# seperate string for the title
title_ = 'Life Expectancy of bigger Asian countries'
```

```python
# using the seperate variables
fig3 = px.bar(df_asia, x = 'country', y = 'lifeExp', 
              color = 'pop', 
              labels = labels_, 
              title = title_)

fig3.show()
```

Apart from customizing our axes labels we can also add data labels to our chart.
Let's do a modified version of the chart to show the life expectancy values on each bar.

```python
# using the text parameter for adding data labels
fig3 = px.bar(df_asia, x = 'country', y = 'lifeExp', 
              color = 'pop', 
              labels = labels_, 
              title = title_,
              text ='lifeExp')

fig3.show()
```

If we want to  display the y axis values as data labels on the bars then we can just use the  ***text_auto*** parameter

```python
# using the text_auto parameter will use the Y axis also for labeling
fig3 = px.bar(df_asia, x = 'country', y = 'lifeExp', 
              color = 'pop', 
              labels = labels_, 
              title = title_,
              text_auto = True
              )

fig3.show()
```

```python
# updating the chart to show the data label next to the bars
fig3.update_traces(textposition = 'outside')

fig3.show()
```

## Stacked bar charts

Stacked bar charts (SBC) show the quantitative relationship that exists between a main categorical variable and its subcategories. Each bar represents a principal category and it is divided into segments representing subcategories of a second categorical variable.

For our stacked bar chart, let's visualize the population distribution of all continents in 2002.

```python
df2 = px.data.gapminder().query('year == 2002')
df2
```

```python
# creating a dictionary for the labels
labels_ = {'continent' : 'Continent', 'pop' : 'Population'}
          
# and a string for the title
title_ = 'Population distribution of continents in 2002'
```

```python
# creating the stacked bar chart
stackedBar = px.bar(df2, x = 'continent', y = 'pop', 
                    color = 'continent',
                    labels = labels_, 
                    title = title_
                    )

stackedBar.show()
```

```python
# let's create a stacked back chart showing the global population growth
df2 = px.data.gapminder()

stackedBar2 = px.bar(df2, x = 'year', y = 'pop', color = 'continent')
stackedBar2.show()
```

```python
# we can also use a histogram chart type which automatically aggregates the data by continent
Histogram1 = px.histogram(df2, x = 'year', y = 'pop', color = 'continent')
Histogram1.show()
```

```python
# we can easily change the histogram to show a 100% percent stacked version 
Histogram1.update_layout(barnorm='percent')
```

#2.**Scatter plots**

We use scatter plots to determine whether or not two variables have a relationship or correlation. The plot uses dots to represent values for the two numeric variables.

## First steps

```python
# creating the scatter plot
fig5 = px.scatter(df_tips, x = 'total_bill', y = 'tip')

fig5.show()
```

## Adding a title

Plotly doesn't generate a title for scatter plots either but to add a title to our plot is just as easy as it was when we did it whilst customizing our bar charts.




```python
title_ = 'Comparing Total bill and Tip values'
```

```python
fig5 = px.scatter(df_tips, x = 'total_bill', y = 'tip',
                  title = title_
                  )

fig5.show()
```

## Changing the coloring and labeling

Adjusting the labels of our scatter plot is also similar to what we did when we were creating bar charts.

```python
# making a seperate dictionary for labels
labels_ = {
           'total_bill' : 'Total Bill',
           'tip' : 'Tip',
           'smoker' : 'Smoker',
           'size' : 'Size'
          } 
```

```python
# creating the scatter plot with labels
fig5 = px.scatter(df_tips, x = 'total_bill', y = 'tip',
                  labels = labels_,
                  title = title_
                 )

fig5.show()
```

Changing the coloring of our scatter plots is simple and similar to how we have handled the coloring of our bar charts.

```python
# coloring with a third column
fig5 = px.scatter(df_tips, x = 'total_bill', y = 'tip',
                  labels = labels_,
                  color = 'smoker',
                  title = title_
                 )

fig5.show()
```

```python
# coloring with a third column
fig5 = px.scatter(df_tips, x = 'total_bill', y = 'tip',
                  labels = labels_,
                  color = 'sex',
                  title = title_
                 )

fig5.show()
```

## Adjusting the size of traces

Since our data set has a 'size' column which represents how many people sat at the specific tables we can use those values to control the size of dots on our scatter plot.

```python
df_tips = px.data.tips()

# using the size parameter to change the dot sizes
fig5 = px.scatter(df_tips.query('size == 2'), x = 'total_bill', y = 'tip', 
                  size = 'size',
                  labels = labels_,
                  title = title_,
                  color = 'smoker'
                  )

fig5.show()
```

We can see that the size column has been automatically added to the hover menu. We can explicitly tell Plotly not to use this column if we want to.

```python
# removing the size column from the hover menu
fig5 = px.scatter(df_tips, x = 'total_bill', y = 'tip', 
                  size = 'size', 
                  hover_data = {'size': False}, 
                  labels =  labels_,
                  title = title_,
                  color = 'smoker'
                  )

fig5.show()
```

## Trendlines

Plotly also supports Linear (OLS) and Non-Linear Trendlines (LOWESS). 

Let's take a look at this by creating the previous chart once with an OLS trendline.

```python
# adding an OLS trendline
fig5 = px.scatter(df_tips, x = 'total_bill', y = 'tip', 
                  size = 'size', 
                  hover_data = {'size': False}, 
                  trendline = 'ols',
                  title = title_,
                  color = 'smoker'
                  )

fig5.show()
```

# 3.Line charts

A line graph â€”also known as a line plot or a line chartâ€” is a graph that uses lines to connect individual data points. A line graph displays quantitative values over a specified time interval.

For our line chart let's use the stocks() sample data set. Each row in this wide dataset represents closing prices from 6 tech stocks in 2018/2019.

## Importing the data

```python
# creating the data frame
df_stocks = px.data.stocks()

# taking a look at the data frame
df_stocks
```

```python
# creating the line chart
fig6 = px.line(df_stocks, x = 'date', y = 'GOOG')

# rendering the chart
fig6.show()
```

## Adding a title

```python
# adding a title the same way we did before
title_ = 'Historical price data for GOOG'
```

```python
fig6 = px.line(df_stocks, x = 'date', y = 'GOOG',
               title = title_
              )

fig6.show()
```

## Adding data point markers

The markers argument can be set to True to show markers on lines.

```python
# using the markers parameter
fig6 = px.line(df_stocks, x = 'date', y = 'GOOG',
               title = title_,
               markers = True
              )

fig6.show()
```

## Customizing the format of hover data

```python
# using the hover data parameter to format shown data
fig6 = px.line(df_stocks, x = 'date', y = 'GOOG',
               title = title_,
               markers = True,
               hover_data = {'GOOG' : ':.3f'}
              )

fig6.show()
```

## Changing the coloring

Let's go back to the gapminder() sample dataset and visualize what the GDP Per Capita Of European countries was over the years that had more than 9.5 million of population.

```python
# creating the data frame
df_eu = px.data.gapminder().query('continent == "Europe" and pop > 9500000')

# taking a look at the first 5 records
df_eu.shape
```

```python
# making a seperate dict for labels
labels_ = {
    'year' : 'Year',
    'gdpPercap' : 'GDP per Capita',
    'country' : 'Country'
}
```

```python
# using the color parameter again
fig7 = px.line(df_eu, x = 'year', y = 'gdpPercap',
               color = 'country',
               hover_data = {'gdpPercap' : ':.2f'},
               markers = True,
               height = 600,
               labels = labels_
              )

fig7.show()
```

We can see that it's hard to differentiate which line belongs to which country with the default coloring. Plotly has several built-in color sequences that we can use.

```python
# adding the color_discrete_sequence parameter
fig7 = px.line(df_eu, x = 'year', y = 'gdpPercap',
               color = 'country',
               color_discrete_sequence = px.colors.qualitative.Light24,
               hover_data = {'gdpPercap' : ':.2f'},
               markers = True,
               height = 300
              )

fig7.show()
```

### Loading the data

We are gonna be using an external data set for this chart called 'Medals_tokyo.xlsx'. This file can be read with the Pandas 'read_excel()' method.

```python
import pandas as pd
```

```python
#reading our excel file into a pandas data frame - local file
#input = pd.read_excel('Medals_tokyo.xlsx')

#reading our excel file into a pandas data frame - remote file
df_input = pd.read_excel('https://biconsulting.hu/download/ceu2022/medals_tokyo.xlsx')

#taking a look at the data
df_input
```

### Top ten bar chart

```python
#filtering and sorting our data frame
df_medals = df_input[df_input.RankTotal < 11]
df_medals = df_medals.sort_values('RankTotal')

#taking a look at our dataset
df_medals
```

```python
#title string
title_= 'Top 10 Teams by total medals'

darks = px.colors.qualitative.Dark24
#creating the bar chart
fig8 = px.bar(df_medals, x = 'Team', y = 'Total',
              title = title_,
              color = 'Team',
              color_discrete_sequence = darks,
              text_auto = True
              )

#additional text customization
fig8.update_traces(textposition = "outside", cliponaxis = False)

fig8.show()
```

```python
px.bar(df_medals, x = 'Team',y ='Total', color ='Team', text_auto = True).update_traces(textposition ="outside")
```

### Grouped bar chart

Let's modify the data frame we used previously to store the 5 best teams by total medals won.

This time we want to display the number medals won by the teams on a grouped bar chart.

Subtasks:
*   Change the filtering of the data frame
*   Create an array for the second level categories
*   Add coloring to the chart based on the medal colors
*   Change the labeling of the chart



#### Sample Chart

image.png

#### Changing the filtering 

```python
#changing the filtering
df_medals = df_medals[df_medals.RankTotal < 6]

df_medals
```

#### The solution

```python
medals = ['Gold', 'Silver', 'Bronze']

#creating the stacked bar chart
groupedBar = px.bar(df_medals, 
                    x = 'Team', 
                    y = medals, 
                    labels = {'variable': 'Medal', 
                              'value' : 'Medal count'},
                    color_discrete_sequence = ['gold', 'silver','#cd7f32'],
                    barmode = 'group',
                    text_auto = True,
                    width = 850,
                    title = 'Medals won by the top 5 teams'
                    )

groupedBar.update_traces(textposition = "outside", cliponaxis = False)

groupedBar.show()
```
