# Faceting

Faceting is the process that split the chart window in several small parts (a grid), and display a similar chart in each section.

It is also known as Trellis or Small multiples charts.

## Simple horizontal faceting

```python
# using the tips dataset
df_tips = px.data.tips()
```

```python
# creating the seperate variables 
labels = {'tip': 'Tip', 'total_bill': 'Total bill', 'sex': 'Gender'}
title = 'Comparing daily tip and total bill values'

# creating an array for the category_orders parameter so that the days are in order
c_orders = ["Thur", "Fri", "Sat", "Sun"]
```

```python
px.scatter(df_tips, x="total_bill", y="tip", color='sex', facet_row="day", 
                         )
```

```python
# this time using the facet_col parameter this way every day gets a new column
scatterFacet2 = px.scatter(df_tips, x="total_bill", y="tip", color='sex', facet_col="day", labels = labels, title = title,
                          category_orders={"day": c_orders}
                         )

scatterFacet2.show()
```

## 2-dimensional faceting

```python
# we can alse use  2 dimensions at the same time
scatterFacet2 = px.scatter(df_tips, x="total_bill", y="tip", color='sex', facet_col="day", facet_row="smoker",
                          category_orders={"day": c_orders}
                         )

scatterFacet2.show()
```

## Faceting for line charts

Faceting is possible with other chart types as well. Let's take a look at a line chart version with the stocks dataset.

```python
df_stocks = px.data.stocks(indexed = True)

df_stocks
```

```python
labels = { 'company' : 'Company', 'date' : 'Date', 'value' : 'Closing Price'  }
```

```python
stocksFacet = px.line(df_stocks, labels = labels, facet_col = 'company')
stocksFacet.show()
```

When the facet dimension has a large number of unique values, it is possible to wrap columns using the facet_col_wrap argument.

```python
stocksFacet = px.line(df_stocks, facet_col = 'company', 
                      facet_col_wrap= 4)

stocksFacet.show()
```

We want to visualize the population distribution of Euorpean countries with over 20 million of population on a faceted bar chart. We want to add coloring to the chart based on life expectancy.

# Animation

Several Plotly Express functions support the creation of animated figures through the animation_frame and animation_group arguments.

## Our first animation

```python
# reseting the df
df_gapminder = px.data.gapminder()

# using the animation_frame and animation_group parameters
animatedGap = px.scatter(df_gapminder, x="gdpPercap", y="lifeExp", animation_frame="year",
                         size="pop", color="continent", hover_name="country", labels = labels_,
                         size_max=55, range_x=[100,100000], range_y=[25,90])

animatedGap.show()
```

To make the dots more spread out we can set the x axis to a logarithmic scale using the **log_x** parameter.

```python
# setting the x axis to a logarithmic scale
animatedGap = px.scatter(df_gapminder, x="gdpPercap", y="lifeExp", animation_frame="year", animation_group="country",
                         size="pop", color="continent", hover_name="country", labels = labels_, 
                         log_x = True, size_max=55, range_x=[100,100000], range_y=[25,90])

animatedGap.show()
```

## Slowing down the animation

If we want the frames of the animation to be longer then we can set the duration of each frame to 1000 which will set a delay of 1 sec between the frames.

```python
animatedGap.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 1000

animatedGap.show()
```

## Saving multiple images in a loop

```python
# getting the values of the day column
days= df.day.unique()
days
```

```python
for _day in days :
  print (_day)
```

```python
days= df.day.unique()
for _day in days :
  df1= df.query(' day == @_day')
  fig= px.scatter(df1,x='total_bill', y='tip', color='sex', width=600, title=_day)
  fig.show()
```

```python
# creating the days list and an empty placehold list
days= df.day.unique()
filenames=[]

for _day in days :

  # filtering the dataframe
  df1= df.query(' day == @_day')

  # creating the figure
  fig= px.scatter(df1,x='total_bill', y='tip', color='sex', width=600, title=_day)

  # creating the filename
  _file= 'tips_' + _day + '.png'
  print ('Saving to ' + _file)
  fig.write_image( _file)

  # saving the filaname in the list - we will need it later
  filenames.append( _file)

# checking the filename list
print(filenames)
```
